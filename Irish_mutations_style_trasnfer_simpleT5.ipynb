```python
# Read the content of the text file
with open("/content/train-source.txt", "r") as file:
    lines = file.readlines()

# Initialize variables
sentences = []
current_sentence = []

# Process the lines
for line in lines:
    line = line.strip()

    if line == "<s>":
        current_sentence = ["<s>"]
    elif line == "</s>":
        if current_sentence:  # Ensure there's content in the sentence
            current_sentence.append("</s>")
            sentences.append(" ".join(current_sentence))
    else:
        current_sentence.append(line)

# Write formatted sentences to a new file
with open("/content/formatted_sentences.txt", "w") as output_file:
    for sentence in sentences:
        output_file.write(sentence + "\n")
```

```python
# Read the content of the text file
with open("/content/train-source.txt", "r") as file:
    lines = file.readlines()

# Initialize variables
tar_sentences = []
current_sentence = []

# Process the lines
for line in lines:
    line = line.strip()

    if line == "<s>":
        current_sentence = ["<s>"]
    elif line == "</s>":
        if current_sentence:  # Ensure there's content in the sentence
            current_sentence.append("</s>")
            tar_sentences.append(" ".join(current_sentence))
    else:
        current_sentence.append(line)

# Write formatted sentences to a new file
with open("/content/formatted_sentences.txt", "w") as output_file:
    for sentence in tar_sentences:
        output_file.write(sentence + "\n")
```

```python
import pandas as pd

data = {'source_text': sentences, 'target_text': tar_sentences}
df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2)
train_df.shape, test_df.shape
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
model.train(train_df=train_df[:5000],
            eval_df=test_df[:100],
            source_max_token_len=128,
            target_max_token_len=50,
            batch_size=8, max_epochs=3, use_gpu=True)
```
